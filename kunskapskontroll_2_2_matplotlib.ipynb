### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
# # # Det är två olika sätt att platta data i Matplotlib. 
# # # 1. Plotta data med det explicita gränssnittet. 
# # # Med detta skapar man ett figur och axlar explicit, och sedan använder dessa för at skapa och konfigurera grafen. 
# # # Man skapar en fiigur och en Axis med plt.subplots()

# # 2. Plotta data med det impliciita gränssnittet.
# # Fördelar med denna metod är snabbare att skriva om man ska göta enkla plottar.
# # Nackdelar är då mindre kontroll över plotten eftersom vi inte skapar referenser till objekten.

# Sättet som rekomenderas generellt är den första metoden, särsklit om man ska skapa komplexa visualiseringar,
# och jobba med fleraundergrafer då den ger bättre flexibilitet och kontroll över visualiseringar, vilket är 
# # bra för mer avanccerad användning. 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# Figure : En Figure är det övergripande containerobjektet som innehåller hela visualiseringen. 
# Man kan tänka på den som en ritduk där alla grafer, undergrafer, och objekt placeras. Det är den övergripande strukturen som innehåller allt.

# En Figure kan innehålla en eller flera Axes (axlar).
# Axes: En Axes är en uppsättning koordinater där data visualiseras. Det är det område i figuren där själva grafen (linjer, staplar, etc.) ritas upp.

# Det finns ofta flera Axes i en Figure, särskilt om du har flera underdiagram i samma figur.
# Varje Axes har sina egna x- och y-axlar.
# Axis: En Axis är en del av Axes och representerar de individuella x- eller y-axlarna i diagrammet, det vill säga själva linjerna som visar skalan för variablerna.

# Det finns en x-axel och en y-axel i varje Axes. De definierar gränserna och enheterna för data som ritas.
# Artist: I Matplotlib är en Artist ett objekt som representerar en visuell komponent som plottas på figuren. Det kan vara nästan alla grafiska objekt, som linjer, punkter, text eller rektanglar. 
# Allt som ritas på figuren, inklusive axlarna, etiketter och titlar, är "Artists" i Matplotlib.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace (-4, 4)
y = x**2

fig, ax = plt.subplots()
ax.plot (x,y)

ax.set_title ("y = x^2")
ax.set_xlabel ("x")
ax.set_ylabel ("y)")
```

```python
x = np.linspace(-4, 4)
y = x**2

plt.plot(x, y)

plt.title('Plot av y = x^2')
plt.xlabel('x')
plt.ylabel('y')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].scatter(random_data_x, random_data_y)
ax[0].set_title('Scatter Plot')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')

ax[1].bar(names, values)
ax[1].set_title('Bar Plot')
ax[1].set_xlabel('Fruit')
ax[1].set_ylabel('Count')

plt.tight_layout()
```
